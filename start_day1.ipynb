```python
# data science 
# combination of Math ,statistics,data analysis and machine learning
# data science is a field that uses scientific methods, processes, algorithms and systems to extract knowledge
# python ,statistics , math 
# python --> pandas ,numpy ,matplotlib ,seaborn , ml-->sklearn ,dl-->(pytorch ,tensorflow)

# statistics --> descriptive statistics , inferential statistics
# math --> linear algebra ,calculus ,probability 
# job roles --> data analyst ,data scientist ,machine learning engineer ,data engineer
# data science process --> data collection ,data cleaning ,data exploration ,data modeling ,data visualization ,model evaluation ,model deployment

# structure data vs unstructured data
# structured data --> data that is organized in a specific format and can be easily stored and accessed
# unstructured data --> data that does not have a specific format and is not organized in a specific way
# examples of structured data --> excel sheets ,csv files ,sql databases
# examples of unstructured data --> text data ,image data ,audio data ,video data

# data cleaning --> the process of identifying and correcting errors and inconsistencies in data to improve its quality and reliability
# data cleaning steps --> handling missing values ,handling outliers ,handling duplicates ,handling inconsistent data
# handling missing values --> removing rows with missing values ,imputing missing values with mean ,median ,mode
# handling outliers --> removing outliers ,imputing outliers with mean ,median
# handling duplicates --> removing duplicate rows
# handling inconsistent data --> standardizing data formats ,correcting typos

# feature engineering --> the process of creating new features from existing data to improve the performance of machine learning models
# feature engineering techniques --> creating new features from existing features ,creating interaction features ,creating polynomial features
# creating new features from existing features --> creating new features by combining existing features
# creating interaction features --> creating new features by multiplying or dividing existing features
# creating polynomial features --> creating new features by raising existing features to a power

# select model --> the process of choosing the best machine learning model for a given problem
#        |
# train model --> the process of fitting a machine learning model to a training dataset
#        |
# predict --> the process of using a trained machine learning model to make predictions on new data
#        |
# evaluate model --> the process of assessing the performance of a machine learning model using various metrics
#        |
# model deployment --> the process of making a machine learning model available for use in a production environment
#        |
# deployment techniques --> creating an API ,creating a web application ,creating a mobile application

#train data vs test data
# train data --> the dataset used to train a machine learning model
# test data --> the dataset used to evaluate the performance of a machine learning model
# persentage of train data and test data --> 70% train data and 30% test data

#EDA --> exploratory data analysis
# EDA --> the process of analyzing and visualizing data to gain insights and understand the underlying patterns and relationships in the data
# EDA techniques --> summary statistics ,data visualization ,correlation analysis
# summary statistics --> mean ,median ,mode ,standard deviation ,variance
# data visualization --> histograms ,box plots ,scatter plots ,heatmaps
# correlation analysis --> the process of measuring the strength and direction of the relationship between two variables
# correlation coefficient --> a measure of the strength and direction of the relationship between two variables
# correlation coefficient values --> -1 to 1 ,0 means no correlation ,1 means perfect positive correlation ,-1 means perfect negative correlation


```
